## *Create* a data for America Movies

 Lets got to wiki page to get the data links of american movies in year.
 https://en.wikipedia.org/wiki/Lists_of_American_films


```python
list_links_of_movies = []
for i in range (1980,2021):
  link="https://en.wikipedia.org/wiki/List_of_American_films_of_"+str(i)
  list_links_of_movies.append(link)
print(list_links_of_movies)
```

```python
import urllib.request
from bs4 import BeautifulSoup

#Get the HTML DATA
url = 'https://en.wikipedia.org/wiki/List_of_American_films_of_1980'
html  = urllib.request.urlopen(url)
soup_data = BeautifulSoup(html,'lxml')

# Extract the table data out of it.
table_data = soup_data.find("table", {"class":"wikitable"})
print(table_data)
```

```python
# Once we have table data class as wikitable lets extract the Title and link of the movie for further processing.
table_data = soup_data.findAll("table", class_="wikitable")
for table in table_data:
  all_headers = [th.text.replace("\n","") for th in table.findAll('th')]
  #  print(all_headers)
  # Only check the table which contains title and Produciton company 
  if ('Title' in all_headers and 'Production company' in all_headers):
    rows = table.tbody.find_all('tr')
    for row in rows[1:10]:
      print(row)
      print("===============================")
      movie_title = row.find_all('i')[0].a
      print(movie_title)
      print(movie_title['href'])
      print(movie_title['title'])
      url = 'https://en.wikipedia.org/'+movie_title['href']
      print(url)
      print("===============================")
    
    break

```

```python
## Get all the movie url from 1980 till 2021
all_movie_urls = []
all_movie_titles = []
failed_count=0
for url in list_links_of_movies:
  try:
    html  = urllib.request.urlopen(url)
    soup_data = BeautifulSoup(html,'lxml')
    table_data = soup_data.findAll("table", class_="wikitable")
    for table in table_data:
      all_headers = [th.text.replace("\n","") for th in table.findAll('th')]
      #  print(all_headers)
      # Only check the table which contains title and Produciton company 
      if ('Title' in all_headers and 'Production company' in all_headers):
        rows = table.tbody.find_all('tr')
        for row in rows[1:]:
          try:
            movie_title = row.find_all('i')[0].a
            all_movie_titles.append(movie_title)
            movie_url = 'https://en.wikipedia.org/'+movie_title['href']
            print(movie_url)
            all_movie_urls.append(movie_url)
          except Exception as e:
            failed_count +=1
  except:
      print(url)


print(f"Total number of url for movies :{len(all_movie_urls)}")

print(failed_count)


    #  https://en.wikipedia.org//wiki/Athlete_A   
```

```python
## Lets convert the above code into fuction to get the movie  plot
def getMoviePlot(movie_url):
  try:
    movie_data = urllib.request.urlopen(movie_url) 
    soup_movie = BeautifulSoup(movie_data,'lxml') 
    plot_header_data =  soup_movie.find(id='Plot').find_parent('h2')

    para = plot_header_data.find_next_sibling()
    plot=""
    while para.name=='p':
      plot=plot+para.text
      para=para.find_next_sibling()
    if plot !="":
      return plot
    else:
      return None
  except :
    return None
```

```python
from tqdm import tqdm
extracted_data = []

for movie_name, movie_url in tqdm(zip(all_movie_titles, all_movie_urls)):
  plot=getMoviePlot(movie_url)
  if plot is not None:
    extracted_data.append([movie_name, movie_url,plot])

print(len(extracted_data))
print(extracted_data[0])
```
